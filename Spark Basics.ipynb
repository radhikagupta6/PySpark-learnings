```
from pyspark.sql.types import StringType, StructField, IntegerType, StructType
```

```
data_schema = [StructField('age', IntegerType(), True),
              StructField('name', StringType(), True)]
```

```
final_stru = StructType(fields = data_schema)
```

```
df = spark.read.json('/FileStore/tables/people-2.json', schema = final_stru)
```

```
df.printSchema()
```

```
type(df['age'])
```

```
df.select('age').show()
```

```
type(df.select('age').show())
```

```
df.head(2)[0]
```

```
type(df.head(2)[0])
```

```
# if we want to view any two columns
df.select(['age','name']).show()
```

```
# creating new column, changes aren't permanent on the original dataframe
df.withColumn('double_age',df['age']*2).show()
```

```
df.show()
```

```
# renaming a column
df.withColumnRenamed('age', 'my_new_age').show()
```

```
# spark.sql we'd have to reg the spark df as a temp table for storing it into the catalog 
df.createOrReplaceTempView('people')
```
